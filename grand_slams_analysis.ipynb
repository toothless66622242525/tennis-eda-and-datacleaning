```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('Mens_Tennis_Grand_Slam_Winner.csv')
```

```python
df.head()
```

```python
prize_by_year = df.dropna(subset=["WINNER_PRIZE"]).groupby('YEAR', as_index=False)["WINNER_PRIZE"].mean()
```

```python
prize_by_year.head()
```

```python
sns.lineplot(data=prize_by_year, x="YEAR", y="WINNER_PRIZE")
```

```python
prize_by_year[(prize_by_year['YEAR'] >= 1998) & (prize_by_year['YEAR'] <= 2003)]
```

```python
df[df["YEAR"] == 2001]
```

```python
row_to_drop = df[(df["YEAR"] == 2001) & (df["TOURNAMENT"] == "French Open")]
```

```python
index_to_drop = row_to_drop.index
```

```python
df_cleaned = df.drop(index_to_drop)
```

```python
prize_by_year_cleaned = df_cleaned.dropna(subset=["WINNER_PRIZE"]).groupby('YEAR', as_index=False)["WINNER_PRIZE"].mean()
```

```python
sns.lineplot(data=prize_by_year_cleaned, x="YEAR", y="WINNER_PRIZE")
```

```python
df_cleaned[df_cleaned["YEAR"] == 1990]
```

```python
years_to_check = [2000, 2001, 1999, 1998, 1997, 1996, 1995, 1994, 1993, 1992, 1991]
```

```python
rows_to_drop = df[df["YEAR"].isin(years_to_check) & (df["TOURNAMENT"] == 
                                                     "French Open")] 
```

```python
indices_to_drop = rows_to_drop.index
```

```python
df_cleaned = df.drop(indices_to_drop)
```

```python
prize_by_year_cleaned = df_cleaned.dropna(subset=["WINNER_PRIZE"]).groupby('YEAR', as_index=False)["WINNER_PRIZE"].mean()
```

```python
sns.lineplot(data=prize_by_year_cleaned, x="YEAR", y="WINNER_PRIZE")
```

```python
df_final_cleaned = df[~(df["TOURNAMENT"] == "French Open")]
```

```python
prize_by_year_final = df_final_cleaned.dropna(subset=['WINNER_PRIZE']).groupby('YEAR', as_index=False)['WINNER_PRIZE'].mean()
```

```python
prize_by_year_final.head()
```

```python
sns.lineplot(data=prize_by_year_final, x="YEAR", y="WINNER_PRIZE")
```

```python
df_final_cleaned.head()
```

```python
sns.lineplot(data=df_final_cleaned, x="YEAR", y="WINNER_PRIZE", hue="TOURNAMENT")
```
